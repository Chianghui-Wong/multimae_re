```python
files = list()
def dirAll(pathname):
    if os.path.exists(pathname):
        filelist = os.listdir(pathname)
        for f in filelist:
            f = os.path.join(pathname, f)
            if os.path.isdir(f):
                dirAll(f)
            else:
                dirname = os.path.dirname(f)
                baseName = os.path.basename(f)
                if dirname.endswith(os.sep):
                    files.append(dirname+baseName)
                else:
                    files.append(dirname+os.sep+baseName)

dict = set()
dirAll("/data/imagenet-S/pretrain-50/train/semseg")
for input_file in files:
    img = Image.open('/data/imagenet-S/pretrain-50/train/semseg/n01443537/n01443537_10007.PNG')
    img_torch = TF.pil_to_tensor(img).to(torch.long).squeeze(0).reshape(-1)
    for i in img_torch:
        dict.add(i.item())
    print(max(dict), len(dict))
    
```
